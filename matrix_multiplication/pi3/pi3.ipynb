# Matrix Multiplication (SGEMM) Benchmarks

## Device: Raspberry Pi 3+

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

%matplotlib widget

plt.rcParams["figure.figsize"] = [10.00, 7.00]
plt.rcParams["figure.autolayout"] = True

columns = ["Frequencies", "ROW64", "ROW128", "ROW256", "ROW512", "ROW1024", "ROW2048"]

data = pd.read_csv("pi3_openblas_flops_per_watt.csv", usecols=columns)

fig, ax = plt.subplots()

plt.title('Flops per Watt by Matrix Rows (Raspberry Pi 3, OpenBLAS)')
plt.xlabel("Frequency (MHz)")
plt.ylabel("Flops per Watt")

openblas64 = ax.plot(data.Frequencies, data.ROW64, label = "64 Rows")
openblas128 = ax.plot(data.Frequencies, data.ROW128, label = "128 Rows")
openblas256 = ax.plot(data.Frequencies, data.ROW256, label = "256 Rows")
openblas512 = ax.plot(data.Frequencies, data.ROW512, label = "512 Rows")
openblas1024 = ax.plot(data.Frequencies, data.ROW1024, label = "1024 Rows")
openblas2048 = ax.plot(data.Frequencies, data.ROW2048, label = "2048 Rows")

leg = ax.legend()
plt.show()
```
